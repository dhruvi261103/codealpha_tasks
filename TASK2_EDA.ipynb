```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# Step 2: Load the dataset
df = pd.read_csv('books_data.csv')

# Step 3: Initial Exploration
print("‚úÖ Dataset Preview:")
print(df.head())

print("\nüîé Dataset Info:")
print(df.info())

print("\n‚ùì Missing Values:")
print(df.isnull().sum())

# Step 4: Convert Rating to Numeric
rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}
df['Rating_Numeric'] = df['Rating'].map(rating_map)

# Step 5: Descriptive Statistics
print("\nüìä Descriptive Statistics:")
print(df[['Price (INR)', 'Rating_Numeric']].describe())

# Step 6: Price Distribution
print("\n‚ùì What price range dominates the dataset?")
plt.figure(figsize=(8,5))
sns.histplot(df['Price (INR)'], kde=True, bins=10, color='skyblue')
plt.title('Price Distribution of Books')
plt.xlabel('Price (INR)')
plt.ylabel('Count')
plt.show()

# Step 7: Rating Distribution
print("\n‚ùì Is the dataset balanced in terms of ratings?")
plt.figure(figsize=(6,4))
sns.countplot(x='Rating_Numeric', data=df, palette='viridis')
plt.title('Book Rating Distribution')
plt.xlabel('Rating (1 to 5)')
plt.ylabel('Count')
plt.show()

# Step 8: Boxplot of Price vs Rating
print("\n‚ùì Are there outliers in prices for specific ratings?")
print("‚ùì Do highly rated books tend to be more expensive?")
plt.figure(figsize=(8,5))
sns.boxplot(x='Rating_Numeric', y='Price (INR)', data=df, palette='pastel')
plt.title('Boxplot: Price vs Rating')
plt.xlabel('Rating')
plt.ylabel('Price (INR)')
plt.show()

# Step 9: Correlation Heatmap
print("\n‚ùì What is the correlation between price and rating?")
plt.figure(figsize=(6,4))
sns.heatmap(df[['Price (INR)', 'Rating_Numeric']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Final Summary
print("\n‚úÖ EDA Completed Successfully!")
```
